```python
# 60. Roulette Payouts

import random

def roulette():
    x = random.randint(0,37)
    if x >= 0 and x < 37:
        print("The spin resulted in " + str(x))
    if x == 37:
        print("The spin resulted in 00")
        print("Pay 00")
    if x == 0:
        print("Pay 0")
    if x >= 1 and x < 37:
        print("Pay " + str(x))
    if x in [1,3,5,7,9,12,14,16,18,19,21,23,25,27,30,32,34,36]:
        print("Pay Red")
    if x in [2,4,6,8,10,11,13,15,17,20,22,24,26,28,29,31,33,35]:
        print("Pay Black")
    if x % 2 == 0 and x >1:
        print("Pay Even")
    if x % 2 != 0 and x < 37:
        print("Pay Odd")
    if x > 0 and x < 19:
        print("Pay 1 to 18")
    if x > 18 and x < 37:
        print("Pay 19 to 36")

roulette()
```

```python
# 63. Temperature Conversion Table

def cels_fahr():

    new_list = []

    for x in range(101):
        if x % 10 == 0:
            new_list.append(x)

    for z in new_list:
        print(str(z) + " degrees celsius is equivalent to " + str(round(z * 9/5 + 32)) + " degrees fahrenheit.")

cels_fahr()
```

```python
# 66. Compute a Grade Point Average

def gpa_function():
    x = 0
    total = 0.0
    b = 0.0
    count = 0
    while x != "":
        x = input("Enter grade. ")
        total = total + b
        if x == "A+" or x == "A":
            b = 4.0
        elif x == "A-":
            b = 3.7
        elif x == "B+":
            b = 3.3
        elif x == "B":
            b = 3.0
        elif x == "B-":
            b = 2.7
        elif x == "C+":
            b = 2.3
        elif x == "C":
            b = 2.0
        elif x == "C-":
            b = 1.7
        elif x == "D+":
            b = 1.3
        elif x == "D":
            b = 1.0
        elif x == "F":
            b = 0
        count += 1
    count = count -1
    gpa = total / count
    print("Your GPA is " + str(gpa) + ".")

gpa_function()
```

```python
# Parity Bits

def parity_bits():
    code = 0
    while code != "":
        code = input("Enter 8-bit code. ")
        if code == "":
            break
        elif len(code) != 8:
            print("Invalid entry. ")
        elif len(code) == 8:
            ones = code.count("1")
            if ones % 2 == 0:
                print("Parity bit required: 0")
            else:
                print("Parity bit required: 1")
                
parity_bits()
```
